# Variables

Variables are extremly useful they let us store values (numbers, texts, lists, data tables, or all sorts of "objects") and help us keep track of things while we write code.

```python
# Variables let us store information
weight_kg = 60.0

print("weight_kg:", weight_kg)
```

```python
# We can operate on variables to get new values
weight_lb = 2.2 * weight_kg
print(weight_lb)
```

```python
# we can even do this with 2 variables
merkats_per_kg = 1/.62
print("There are roughly up to", 
      merkats_per_kg*weight_kg, 
      "meerkats in", weight_kg, "kg")
```

```python
# We can also use the math package to do math
import math

2 * math.log2(10)
```

```python
# We can store all kinds of things in variables
color = "blue"
list1 = [3,4,5,8,"purple"]

print("color:", color)
print("list1:", list1)
```

## Strings: a kind of object

```python
# strings store text
sequence = "ATCG"
```

```python
# You can get out individual letters like this
print(sequence[0])
print(sequence[3])
```

```python
# Or whole chunks of the string like this
print(sequence[0:3])
print(sequence[:3])
print(sequence[3:])
```

```python
# you can also add other strings to strings
print(sequence + "A")

# or "multipy" strings
print(sequence * 2)
```

```python
# there are a lot of other things you can do to 
# strings with default python

print(sequence.replace("A", "C"))
```

```python
# We can also use things like the len function to get 
# information about a string

print(len(sequence))
```

# Lists

Lists help us organize and store multiple objects for later use.

**A list is not saved unless you save it to a variable!**


```python
# example list
favorite_numbers = [23, 3, 8]
print(favorite_numbers)
```

```python
# We can acess individual entries in a list
print(favorite_numbers[0])
print(favorite_numbers[2])
print(favorite_numbers[-1])
```

```python
# You can add entries to lists with append
print("favorite_numbers before append:", favorite_numbers)
print(favorite_numbers)
favorite_numbers.append(2)
print("favorite_numbers after append:", favorite_numbers)

# remove entries with remove
print("favorite_numbers before remove", favorite_numbers)
favorite_numbers.remove(3)
print("favorite_numbers after remove", favorite_numbers)

# or add multiple entries with extend
print("favorite_numbers before extend", favorite_numbers)
favorite_numbers.extend([12, 17])
print("favorite_numbers after extend", favorite_numbers)
```

```python
# lists can contain numbers and strings (or any other objects)
l1 = ["blue", 5]
```

# Dictonaries

```python
# Dictionaries allow us to store values with keys
colors = {"red":"mad", 
          "blue":"sad", 
          "yellow":"happy"}

print(colors["red"])
print(colors["yellow"])
```

### Setting up functions:

Next we are going to make the function draw_snowman(). I already wrote the code for this so you just need to run the next cell to get it up and running, then you can do the bonus exercise. This function will make a snowman you specify.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches


def draw_snowman(snowman_x, snowman_y, snowman_size, title="My Snowman", figsize=(6, 8)):
    """Draw a snowman with any number of circles, with arms on the second circle if present."""

    # Make sure all three lists are the same length
    assert len(snowman_x) == len(snowman_y), (
        f"Every circle needs both an X and a Y coordinate — looks like you have "
        f"{len(snowman_x)} X value(s) but {len(snowman_y)} Y value(s) in snowman_y. "
        f"Looks like part of your snowman is floating in a mysterious void! Make sure you have the same number of X and Y values"
    )
    assert not (len(snowman_x) > len(snowman_size)), (
        f"Every circle needs a size — looks like you have "
        f"{len(snowman_x)} circle(s) but {len(snowman_size)} size(s) in snowman_size "
        f"Every snowball needs a size even if it is small"
    )
    assert not (len(snowman_x) < len(snowman_size)), (
        f"Every circle needs both an X and a Y coordinate "
        f"{len(snowman_x)} circle(s) but {len(snowman_size)} size(s) in snowman_size. "
        f"Looks like part of your snowman is floating in a mysterious void! Give it coordinates"
    )

    fig, ax = plt.subplots(figsize=figsize)

    # Draw however many circles were passed in
    for i in range(len(snowman_x)):
        circle = plt.Circle(
            (snowman_x[i], snowman_y[i]),
            snowman_size[i],
            color='white', ec='black', lw=2, zorder=2
        )
        ax.add_patch(circle)

    # Draw arms on the second circle (index 1) only if it exists
    if len(snowman_x) > 1:
        arm_scale = snowman_size[1] / 0.7
        ax.plot(
            [snowman_x[1] - snowman_size[1],  snowman_x[1] - snowman_size[1] - 0.9 * arm_scale],
            [snowman_y[1] + 0.15 * arm_scale, snowman_y[1] + 0.6 * arm_scale],
            color='saddlebrown', lw=3, solid_capstyle='round', zorder=1
        )
        ax.plot(
            [snowman_x[1] + snowman_size[1],  snowman_x[1] + snowman_size[1] + 0.9 * arm_scale],
            [snowman_y[1] + 0.15 * arm_scale, snowman_y[1] + 0.6 * arm_scale],
            color='saddlebrown', lw=3, solid_capstyle='round', zorder=1
        )

    ax.set_xlim(-4, 4)
    ax.set_ylim(-1.5, 5)
    ax.set_aspect('equal')
    ax.set_facecolor('#daeef7')
    ax.set_xticks(range(-4, 5))
    ax.set_yticks(range(-1, 6))
    ax.grid(True, linestyle='--', alpha=0.35)
    ax.set_title(title, fontsize=14, pad=12)

    plt.tight_layout()
    plt.show()
```

## Bonus problems:

Using the code snippit below try to do the following:

- Adjust the overall size of the snowman by changing the *scale* variable
- Have the head roll off the snowman and onto the ground by adjusting the *snowman_x* and *snowman_y* variables (by changing the entries in the lists)

```python
# Example Snowman

# Variable (storing a number)
## This variable is the "scaling factor" for the snowman
## increasing it will lead to an overall larger snowman
scale = 1.0



# Variable (storing a list)
## This is a list of the X positions of the diffrent parts 
## (snowballs) of the snowman.
snowman_x = [0, 0, 0]




# Variable (storing a list)
## This is a list of the Y positions of the diffrent parts 
## (snowballs) of the snowman.
## Note that we use scale here again to make sure we adjust
## the distance between snowballs so they stay (realtivly)
## the same as the snowman gets bigger.
snowman_y = [0,
             (1.0 + 0.7) * scale,
             (1.0 + 0.7 + 0.7 + 0.5) * scale]


# Variable (storing a list)
## This is a list of the sizes of the diffrent parts 
## (snowballs) of the snowman.
snowman_size = [1.0 * scale,
                0.7 * scale,
                0.5 * scale]



# ── Draw your result ──────────────────────────────────────────────────────────
draw_snowman(snowman_x, snowman_y, snowman_size)
```
